```python
import numpy as np
import pandas as pd          
df=pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv",sep=',')
df.head()
df.shape
```

```python
df.head()
```

```python
df.drop(df.columns[df.columns.str.contains('unnamed', case=False)],
          axis=1, inplace=True)                                            #1 question
# dropping the column which is unnamed. 
```

```python
df.head()   # to display the first few rows of data frame
```

```python
df["Gender"].value_counts()                                               # 2 question

# returns series of unique value
```

```python
top = df.groupby("Name").sum()                                              #3 question
top.sort_values("Count",ascending=False).head()           # grouping based on name and sorting the count in descending order
```

```python
grouped=df.groupby("Name")                   # 4 question
grouped["Count"].agg(np.median)              #median occurence
                                             # group by name and taking its median
```

```python
grouped=df.groupby(["State","Gender"])  #5 question
print(grouped["Count"].agg(np.sum)      # grouping based on state and gender and sum of counts over states
     )
```

```python
grouped= df.groupby("Gender")
print(grouped["Count"].agg(np.sum))                      
```
